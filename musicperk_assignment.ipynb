```python
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2017,1,1)
end = datetime.datetime(2019,1,1)
```

```python
facebook = web.DataReader('FB','yahoo',start,end)
```

```python
facebook.head()
```

```python
facebook['Open'].plot(figsize = (16,6))
```

```python
df20d = facebook.rolling(20, center=True).mean()
df20d.head(10)
df50d = facebook.rolling(50, center=True).mean()
df50d.head(10)
```

```python
facebook['Open'].plot(figsize=(16,6))
df20d['Open'].plot(label='Facebook 20 Day SMA',color = 'orange')
df50d['Open'].plot(label='Facebook 50 Day SMA',color = 'magenta')
plt.legend(loc='upper left')
fig = plt.figure()
fig.savefig('test.png')
```
